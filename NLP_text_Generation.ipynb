First, you will create a function which takes a list of words, and returns a dictionary of `nextwords`. This is similar to the dictionary you created in the last homework assignment:  keys are items in the list of words, and corresponding values are the list of all words that follow them.

This time, however, there's a twist. We'll add a special key to the dictionary, which will be the symbol '%', signifying the first word of a sentence. When scanning the wordlist to construct the `nextwords` dictionary, normally you would add `wordlist[i+1]` to the list `nextwords[wordlist[i]]`. However, if `wordlist[i]` ends with a period, then you should instead add `wordlist[i+1]` to the list for nextwords['%']. (Make sure you also add the very first word of `wordlist` to nextwords['%'].)


```
> shakespeare_nextwords=CreateNextWords(shakespeare_words)
> shakespeare_nextwords['Rose']
['at', 'or', 'or', 'and']

> len(shakespeare_nextwords['%'])
76543
```

````python
def CreateNextWords(wordlist):
  nextwords={}
  nextwords['%']=[wordlist[0]]
  for word in range(len(wordlist)-1):
    if wordlist[word] not in nextwords.keys():
      nextwords[wordlist[word]]=[]
    if wordlist[word][-1]=='.':
      nextwords['%']+=[wordlist[word+1]]
    else:
     nextwords[wordlist[word]]+=[wordlist[word+1]]
  return nextwords
````

We now move on to text generation.  First, we'll need the choice function:

````python
from random import choice
````

Now, create a function called `MakeNewPlay(nextwords,n)` which takes a dictionary `nextwords` and a number `n`, and follows these steps:
1. Define an empty `newplay` string.
2. Set `currentword` equal to '%'.
3. Start a for-loop that will run `n` times.
4. Randomly choose a word from the list `nextwords[currentword]`.
5. Add the chosen word to `newplay`, followed by a space `' '` character.
6. If the chosen word ends in a period, add a newline character '\n' and set `currentword` equal to `%`.
7. Otherwise, set `currentword` equal to the chosen word.
8. Go back to step 4.
9. When all `n` iterations have finished, your function should return `newplay`.

````python
def MakeNewPlay(nextwords,n):
  newplay=''
  currentword='%'
  for word in range(n):
    result=choice(nextwords[currentword])
    newplay+=result+' '
    if result[-1]=='.':
      newplay+='\n'
      currentword='%'
    else:
      currentword=result
  return newplay
````

To create a 1000 word Shakesperian play, run the following code block:

````python
nextwords=CreateNextWords(shakespeare_words)
newplay=MakeNewPlay(nextwords,1000)
print(newplay).
````
